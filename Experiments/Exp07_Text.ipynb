```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('demon.png')

# Add text
cv2.putText(img,'DEMON',(50,100),cv2.FONT_HERSHEY_SIMPLEX, 2, (0,255,0), 3)

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display image
plt.imshow(img_rgb)
plt.title("Text on Image")
plt.axis('off')
plt.show()
```
