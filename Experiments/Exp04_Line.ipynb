```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('demon.png')

cv2.line(img, (50, 50), (500, 50), (255, 0, 0), 3)
cv2.line(img, (50, 100), (500, 100), (0, 255, 0), 3)
cv2.line(img, (50, 150), (500, 150), (0, 0, 255), 3)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis('off')
plt.show()
```
