```python
import cv2
import matplotlib.pyplot as plt

# Read images
img1 = cv2.imread(r"C:\Users\V.Siddharth\OneDrive\Pictures\demon.png")
img2 = cv2.imread(r"C:\Users\V.Siddharth\OneDrive\Pictures\luffg.png")

# Check if images loaded successfully
if img1 is None:
    print("Error: demon.png not found")
elif img2 is None:
    print("Error: luffg.png not found")
else:
    # Resize second image to match first image
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

    # Blend images
    blended = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

    # Convert BGR to RGB for matplotlib
    img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    blended_rgb = cv2.cvtColor(blended, cv2.COLOR_BGR2RGB)

    # Display results
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.imshow(img1_rgb)
    plt.title("Image 1")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(img2_rgb)
    plt.title("Image 2")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.imshow(blended_rgb)
    plt.title("Blended Image")
    plt.axis("off")

    plt.show()
```
