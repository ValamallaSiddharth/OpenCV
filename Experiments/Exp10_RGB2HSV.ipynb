```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('demon.png')

# Convert BGR to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Display Original Image
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

# Display HSV Image
plt.subplot(1,2,2)
plt.imshow(hsv)
plt.title("HSV Image")
plt.axis('off')

plt.show()
```
