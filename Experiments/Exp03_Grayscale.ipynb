```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('demon.png')

# Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display Grayscale Image
plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.show()
```
