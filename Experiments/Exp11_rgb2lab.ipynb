```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("demon.png")

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

lab_img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(rgb_img)
plt.title("Original RGB")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(lab_img)
plt.title("LAB Image")
plt.axis('off')

plt.show()
```
