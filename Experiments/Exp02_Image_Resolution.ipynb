```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('demon.png')

# Get resolution
height, width, channels = img.shape

print("Height :", height)
print("Width  :", width)
print("Channels :", channels)

# Display image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title(f"Resolution: {width} x {height}")
plt.axis('off')
plt.show()
```
